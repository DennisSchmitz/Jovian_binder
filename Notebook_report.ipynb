# READ THIS FIRST
**This is an <u>example</u> notebook containing an amalgamation of publicly available examples and simulated data. Please keep in mind that graphs shown here do not always exactly represent the graphs as generated by Jovian. The reason why we have done this is because we are using Binder to show you this report. Binder hosts this report for free but requests to keep total file sizes low (<10MB). Since some of the graphs generated by Jovian are larger than this, we show you public examples as a replacement. When we show you these public examples, above it you'll find an explanation of the information shown when generated via Jovian.**

### Instructions:  
- Data should be automatically loaded when you first start this notebook, however, it can take a couple of minutes (depending on the size of your dataset). As long as the circle in the top right corner is not-empty (i.e. completely filled), the data is still being loaded. If something went wrong and you want to manually visualize all the data, first click the "`Cell`" button on the top left corner of this website and then clicking "`Run all`".

### Read-based composition of analyzed samples:

**Low-quality** reads are those that did not meet the stringency settings as specified in the config file.  
**Unclassified** reads are those that could not be assigned to a taxa.  
**Remaining** reads are those that could not be assembled into contigs longer than the user specified minimum contig length.
