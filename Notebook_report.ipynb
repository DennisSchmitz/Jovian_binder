# Read this first

**This is an <u>example</u> notebook containing an amalgamation of publicly available data and technical examples. Please keep in mind that data shown here does not always exactly represent the data as generated by Jovian.**  

The reason why we have done this is because we are using Binder to show you this report.  
Binder hosts this report for free but requests to keep total file sizes low (<10MB). Most of the data which is generated by Jovian is larger than this. Therefore, this notebook shows minimalized data from a public dataset and/or public examples as a replacement of said data.


**Above every displayed result in this report, you'll find a quick explanation of what is being shown.
We have actively tried to construct this example-notebook to be a near-copy of a real Jovian report.**

### Instructions:  
**- Data should be automatically loaded when you first start this notebook, however, it can take a couple of minutes (depending on the size of your dataset). ## Quality control metrics report (MultiQC):
[Open MultiQC graph in separate tab by clicking here](results/multiqc.html)  
___

The Quality Control metrics report is an aggregate report of various bioinformatics tools. In the case of Jovian, several statistical sets are shown:
- General statistics of the input samples
- Number of reads with a given insert size for every sample, done with Picard
- Number of reads that were aligned with a given reference, done with Bowtie2
- Number of trimmed reads, done with Trimmomatic
- Additional Quality Control statistics provided by FastQC

The samples of this report can be found in the following, publicly available, dataset: https://www.ncbi.nlm.nih.gov/bioproject/?term=PRJNA491626

### Read-based composition of analyzed samples:
[Open the barchart in separate tab by clicking here](data/Sample_composition_graph.html)  
<br>
The Read-based composition barchart gives you a quick overview of the contents of each sample processed by Jovian. 
<br>
<br>
**Low-quality** reads are those that did not meet the stringency settings as specified in the config file.  
**Unclassified** reads are those that could not be assigned to a taxa.  
**Remaining** reads are those that could not be assembled into contigs longer than the user specified minimum contig length.
